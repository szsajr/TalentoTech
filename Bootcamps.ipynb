######Histograma: Numero de asesinato(por cada 100k habitantes)

```python
def graph_by_state(col, estimator='mean'):
    if estimator == 'mean':
        state = data.groupby('state')[col].mean().reset_index()
    elif estimator == 'median':
        state = data.groupby('state')[col].median().reset_index()
    else:
        print('estimator not recognized')
    state.sort_values(by=col, ascending=False, inplace=True)

    plt.figure(figsize=(12,12))
    sns.barplot(data=state, y='state', x=col, palette='winter')
    plt.show()
```

```python
graph_by_state('murdPerPop', estimator='median')
```

Podemos observar que el estado de Washington DC presenta el mayor al mayor número de asesinatos(por cada 100k habitantes), mientras que el estado de Illinois presenta el menor número de asesinatos(por cada 100k habitantes).
